### Mensagens

Utilizaremos 3 tétnicas para criar mensagens customizadas:
- Concatenação de Strings
- F-Strings
- T-Strings

### Concatenação de Strings (+)

```python
# Exemplo 1

# nome.sobrenome@senai.com

nome = 'jorginho'
sobrenome = 'moura'

# 'joao.silva@senai.com'

print(nome + '.' + sobrenome + '@senai.com')
print()

# Exemplo 2

codigo_produto = "#10"
quantidade_produto = 100
preco_produto = 3.9

# Código: #10, Quantidade: 100 unidades, Preço: R$ 3.90

print("Código: " + codigo_produto + ", Quantidade: " + str(quantidade_produto) +
     ", Preço: R$ " + str(preco_produto))
```

### F-Strings

```python
# Exemplo 1

nome = 'joao'
sobrenome = 'silva'

# joao.silva@senai.com

email = f"{nome}.{sobrenome}@senai.com"
print(email)

# Exemplo 2

codigo_produto = "#10"
quantidade_produto = 100
preco_produto = 3.9

# Código: #10, Quantidade: 100 unidades, Preço: R$ 3.90

print(f"Código: {codigo_produto}, Quantidade: {quantidade_produto}, "
      f"Preço: R$ {preco_produto}")
print()

# Com formatador
print(f"Código: {codigo_produto}, Quantidade: {quantidade_produto}, "
      f"Preço: R$ {preco_produto:.2f}")
```

```python
# Formatadores de F-Strings

# Com formatador de casas decimais
print(f"Código: {codigo_produto}, Quantidade: {quantidade_produto}, "
      f"Preço: R$ {preco_produto:.2f}")
print()

# Formatador para separar os milhares
quantidade_vendida = 3_000_000
print(f"Quantidade vendida no ano foi de: {quantidade_vendida:,.2f}")
print()

# Formatador de porcentagem (%)
valor = 1000
desconto = .3

print(f"Valor: R$ {valor:.2f}")
print(f"Desconto: {desconto:.2%}")
print(f"Valor com desconto: {valor * (1 - desconto):.2f}")
```
