#### Regressão Linear

É a técnica para criação de modelos preditivos com base no Métodos dos Mínimos Quadrados Ordinários (OLS). Que visa gerar uma reta de forma que a aproximação gere o menor erro.
<br>
Quanto menor o erro, melhor o modelo.
<br>
Um modelo pode ter n variáveis independentes, porém devem ser escolhidos cuidadosamente, para que não atrapalhem ao invés de ajudar.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Criando uma pequena base de dados
data = {
    'area': [50, 70, 90],
    'preco': [250_000, 350_000, 450_000 ]
}

# Transformar os dados em DataFrame
df = pd.DataFrame(data)
```

```python
df.info()
```

```python
df.describe()
```

```python
# Determinando a variável independente X
X = df[['area']]

# Determinar a variável dependente y
y = df['preco']

# Instanciando um objeto LinearRegression para informarmos as variáveis envolvidas
model = LinearRegression()

# O próprio modelo criará a equação de reta
model.fit(X, y)
```

```python
# Plotar o gráfico para vermos a aproximação feita pelo modelo
plt.scatter(X, y, color='b')
plt.plot(X, model.predict(X), color='orange')
plt.title('Preço do terreno x área em m²')
plt.xlabel('Área do Terreno em m²')
plt.ylabel('Preço do terreno em R$')
plt.show()
```

```python
# Montando e usando o modelo

area_previsao = 100
preco_previsto = model.predict([[area_previsao]])
print(f'O preço previsto para um terreno de {area_previsao} m²'
      f' é de R$ {float(preco_previsto):,.2f}')
```

```python
print('Intersepto', model.intercept_)
print('Coeficiente de X', model.coef_)

# y = 0 + x1.5000
```
