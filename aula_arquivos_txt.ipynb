### Função open()

Utilização: Sepre que precisarmos abrir um arquivo, seja .txt, .png entre outros.

A função open() recebe dois arguentos, 'nome_arquivo.extensao' e o 'modo'

**Modos:**
- 'w' de write: Para escrever
- 'r' de read: Para ler
- 'a' de append: Para adicionar conteúdo

```python
file = open('teste.txt', 'r')
print(file.read())

file.close()
```

```python
# Abrindo para escrita
with open('dados_1.txt', 'w', encoding='utf-8') as file:
    file.write('Nome: João Silva')

# Abrindo para leitura
with open('dados_1.txt', 'r', encoding='utf-8') as file:
    print(file.read())
```

```python
# Abrindo para adicionar conteúdo
# ⚠️ Cuidado para não apagar o que já tinha no arquivo!

with open('dados_1.txt', 'a', encoding='utf-8') as file:
    file.write('\nE-mail: joao.silva@senai.com')
    file.write('\nData de Nascimento: 10/10/1990')
```

```python
dados = {}

with open('dados.txt', 'r', encoding='utf-8') as file:
    for linha in file.readlines():
        campo = linha.split(':')[0]
        valor = linha.split(':')[1].strip().replace('\n', '')
        dados[campo] = valor

print(dados)
```
