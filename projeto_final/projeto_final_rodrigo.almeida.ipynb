```python
# Importando Bibliotecas

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Extraindo a base de dados

df = pd.read_csv("car_prices.csv").copy()

# Metadados do arquivo

df.info()
```

```python
# Limpando as linhas que não possuem valores

df = df.dropna() # --> Dropna: Essencial para remover linhas e colunas com valores nulos (Nan ou None)
df.info()
```

```python
# Limpando as linhas duplicadas

df = df.drop_duplicates()
df.info()
```

```python
# Filtrando o Dataframe para que só apresente os registros que sejam do ano 2010

df_2010 = df.loc[lambda df: df["year"] == 2010] # --> loc eele

# Metadados do arquivo

df_2010.info()
```

```python
# Apresentando as primeiras cinco linhas da planilha

df_2010.head()
```

```python
# Listando as marcas mais vendidas por ordem de unidade

marca_mais_vendida = df_2010.value_counts(["make"])
print(marca_mais_vendida)
```

```python
# Extraindo as 5 primeiras marcas listadas pelo value_counts e deixando como valor na constante top_five_marca

top_five_marca = marca_mais_vendida.head()
top_five_marca
```

```python
# Gerando o gráfico

fig, ax = plt.subplots(figsize=(15, 5))
top_five_marca.plot(kind = "bar", color =sns.mpl_palette ("Pastel1"), edgecolor = "black")


plt.title("Top 5 das marcas mais vendidas do ano de 2010 ")
plt.xlabel("Marcas")
plt.ylabel("Unidades vendidas")
plt.xticks(rotation = 0)

ax.grid()
plt.show()
```

```python
# Normalizando os caracteres da coluna body para mínuscula e listando os bodies mais vendidos por ordem de unidade

df_2010["body"] = df_2010["body"].str.lower()

body_mais_vendido = df_2010.value_counts(["body"])
print(body_mais_vendido)
```

```python
#  Extraindo as 5 primeiras marcas listadas pelo value_counts e deixando como valor na constante top_five_body

top_five_body = body_mais_vendido.head()
top_five_body
```

```python
# Gerando o gráfico

fig, ax = plt.subplots(figsize=(15, 5))
top_five_body.plot(kind = "barh", color =sns.mpl_palette ("Pastel1"), edgecolor = "black")


plt.title("Top 5 dos bodies mais vendidas do ano de 2010 ")
plt.xlabel("Unidades vendidas")
plt.ylabel("Body")
plt.xticks(rotation = 0)

ax.grid()
plt.show()
```

```python
# Listando as financiadoras que realiazaram mais financiamentos no ano de 2010

financiadoras = df_2010.value_counts(["seller"])
print(financiadoras)
```

```python
# Extraindo as 5 primeiras financiadoras listadas pelo value_counts e deixando como valor na constante top_five_financiadora

top_five_financiadora = financiadoras.head()
top_five_financiadora
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
top_five_financiadora.plot(kind = "bar", color =sns.mpl_palette ("Pastel1"), edgecolor = "black")


plt.title("Top 5 das financiadoras que realizaram mais financiamentos no ano de 2010 ")
plt.xlabel("Financiadoras")
plt.ylabel("Financiamentos realizados")
plt.xticks(rotation = 1)

ax.grid()
plt.show()
```
