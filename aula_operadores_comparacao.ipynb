### Operadores de Comparação

Utilização: Sempre que precisarmos comparar valores e/ou <br>
elaborar estruturas condicionais.<br>

**São eles:**

- == ➡️ Igual a
- != ➡️ Diferente de
- \> ➡️ Maior que
- < ➡️ Menor que
- \>= ➡️ Maior ou igual a
- <= ➡️ Menor ou igual a

Com esses operadores, montamos comparações e/ou expressões lógicas<br>
cujos resultados sempre são ou verdadeiro (*True*) ou falso (*False*)

```python
# Igual a

print(10 == 10)
print(10 == 10.0)
print(10 == '10')
print(10 == 2 * 5)
print('admin' == 'admin')
print('admin' == 'Admin')
print('site.com' == 'site.rn')
```

```python
# Diferente de 

print(10 != 10)
print(10 != 10.0)
print(10 != '10')
print(10 != 2 * 5)
print('admin' != 'admin')
print('admin' != 'Admin')
print('site.com' != 'site.rn')
```

```python
# Maior que

print(10 > 10)
print(10 > 2)
print(10 > 20)
print(10 > 100 / 5 / 2)
```

```python
# Menor que

print(10 < 10)
print(10 < 2)
print(10 < 20)
print(10 < 100 / 5 / 2)
```

```python
# Maior ou igual a

print(10 >= 10)
print(10 >= 2)
print(10 >= 20)
print(10 >= 100 / 5 / 2)
print(len('oi') >= len('tchau'))
```

```python
# Menor ou igual a

print(10 >= 10)
print(10 >= 2)
print(10 >= 20)
print(10 >= 100 / 5 / 2)
print(len('oi') >= len('tchau'))
```

```python
# Exemplo 1

media = 10
status_aluno = media >= 5
print(f"Com méida de {media} o aluno aprova? {status_aluno}")
```
