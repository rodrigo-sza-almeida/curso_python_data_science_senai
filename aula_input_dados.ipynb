### Input de Dados

Utilização: Sempre que quisermos simular alguma interação com o usuário.

- Ajuda a testar a aplicação e o tipo de dado a ser recebido/tratado
- Função input()

```python
# Exemplo 1

nome = input("Informe seu nome: ")
sobrenome = input("Informe seu sobrenome: ")

email = f"{nome}.{sobrenome}@senai.com"
print(email)
```

```python
# Exemplo 2

codigo_produto = input("Código do produto: ")
quantidade_produto = int(input("Quantidade do produto: "))
preco_produto = float(input("Preço do produto: R$ "))

print(f"Código: {codigo_produto}, Quantidade: {quantidade_produto},"
      f"Preço: R$ {preco_produto:.2f}")
```

```python
print(type(codigo_produto))
print(type(quantidade_produto))
print(type(preco_produto))
```
