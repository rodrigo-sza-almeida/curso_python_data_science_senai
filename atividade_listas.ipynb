```python
from statistics import mean

print('Registro de Atletas - Salto a Distância 🏃‍➡️')

while True:

    nome =input('Nome: ').title()
    if nome:
        saltos = []
        for i in ['Primeiro', 'Segundo', 'Terceiro', 'Quarto', 'Quinto']:
            salto = float(input(f'{i} salto: '))
            saltos.append(salto)

        media = mean(saltos)
        
        print('--- Resultado final ---')
        print(f'Nome do Atleta: {nome}')
        print('Saltos:', ' - '.join(map(str, saltos)))
        print(f'Média dos saltos: {media} m')
        
    else: 
        break
```
