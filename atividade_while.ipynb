```python
# Sistema de Banco

saldo = 0

print('--- Sistema Banco Senai 🏦 ---')

while True:

    print(f'Seu saldo final é de R$ {saldo:,.2f}')
    
    print('''
    Escolha a operação que deseja:
    [ 1 ] -> Depósito
    [ 2 ] -> Saque
    [ 3 ] -> Sair
    ''')
    operacao = input(':')
    
    if operacao == '3':
        print('Volte sempre ...')
        break # para sair do loop ❌

    elif operacao == '1':
        print('--- Depósito ---')
        valor = float(input('Valor a ser depositado: R$ '))
        saldo += valor  # incrementando de forma elegante 👌

    elif operacao == '2':
        print('--- Saque ---')
        valor = float(input('Valor a ser sacado: R$ '))
        if valor <= saldo:
            saldo -= valor # decrementando de forma elegante 👌
        else:
            print('Saldo insuficiente!')
            continue
    
    else:
        print('Operação inválida!')
        continue
print(f'Seu saldo final é de R$ {saldo:,.2f}')
        
        
    
```

### Usando Match / Case

```python
# Sistema de Banco

saldo = 0

print('--- Sistema Banco Senai 🏦 ---')

while True:

    print(f'Seu saldo final é de R$ {saldo:,.2f}')
    
    print('''
    Escolha a operação que deseja:
    [ 1 ] -> Depósito
    [ 2 ] -> Saque
    [ 3 ] -> Sair
    ''')
    operacao = input(':')

    match operacao:
    
        case '3':
            print('Volte sempre ...')
            break # para sair do loop ❌
    
        case '1':
            print('--- Depósito ---')
            valor = float(input('Valor a ser depositado: R$ '))
            saldo += valor  # incrementando de forma elegante 👌
    
        case'2':
            print('--- Saque ---')
            valor = float(input('Valor a ser sacado: R$ '))
            if valor <= saldo:
                saldo -= valor # decrementando de forma elegante 👌
            else:
                print('Saldo insuficiente!')
                continue
        
        case _:
            print('Operação inválida!')
            continue
print(f'Seu saldo final é de R$ {saldo:,.2f}')
        
        
    
```
