### Dicionarios Python 🗂️

Utilização: Sempre que precisarmos de uma coleção que tenha uma estrutura do tipo chave/valor

**Características:**
- Representação: {chave1: valor1, chave2: valor2, ..., chaveN: valorN}
- São iteráveis
- ❌Não são indexadas em zero 
- Podem ser heterogêneas
- São editáveis ✏️
- Pertencem a classe dict

**Vantagens**
- Armazenamento de n informações
- Auxiliam na organização de dados
- Compatíveis com JSON
- Compatíveis com Bancos de dados NoSQL do tipo documento, por exemplo MongoDB

```python
usuario = {
    'id': 1,
    'nome': 'João Silva',
    'email': 'joao.silva@senai.com',
    'data_nascimento': '10/10/1995'
}

print(usuario)
print(type(usuario))

for item in usuario:
    print(item)
```

### CRUD usando as chaves

```python
# Adicionar um par
usuario['endereco'] = 'Rua do Senai 100'
print(usuario)

# Ler um valor
print(usuario['nome'])

# Atualizar um par
usuario['endereco'] = 'Rua do Senai 200'
print(usuario)

# Deletar um par
del  usuario['endereco']
print(usuario)
```

### CRUD usando os métodos da classe

```python
# Adicionar um par
usuario.update(endereco='Rua do Senai 100')
print(usuario)

# Ler um valor
print(usuario.get('nome'))

# Atualizar um par
usuario.update(endereco='Rua do Senai 200')
print(usuario)

# Deletar um par
usuario.pop('endereco')
print(usuario)
```

### Métodos para extração das chaves, valores ou ambos


```python
# quero apenas as chaves
campos = usuario.keys()
for campo in campos:
    print(campo)

print()

# quero apenas os valores
valores = usuario.values()
for valor in valores:
    print(valor)

print()

# quero chavres e valores 

for chave, valor in usuario.items():
    print(chave, valor)
```
