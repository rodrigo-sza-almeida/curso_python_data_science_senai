### Loop for

Utilização: Sempre que precisarmos repetir uma dada estrutura de código baseado em um número, seja uma quantidade de letras em uma string, quantidade de elementos em uma lista ou até mesmo baseado em um número inteiro.

**Vantagens:**
- Otimização de código;
- Otimização no processamento de dados;
- Maior controle de repetição no código;

**Sintaxe:**

<code>for i in n_iteracoes:</code><br>
<code>    bloco do for</code>

```python
texto = 'Python para Data Science'

for letra in texto:
    print(letra, f'-> Código ASCII: {ord(letra)}')

print(len(texto))
```

```python
alunos = ['João', 'Ana', 'Daniel', 'Maria'] 
print('Quantidade de alunos:', len(alunos))

for aluno in alunos:
    n1 = float(input("N1: "))
    n2 = float(input("N2: "))
    
    # Cálculo da média
    media = (n1 + n2) / 2
    print()
    print(f"{aluno}, média é de {media:.2f}")
    
    # Condicional
    if media == 10:
        print("Aprovado(a) com distinção!")
    
    elif 10 > media >= 6:
        print("Aprovado(a)!")
    
    elif 6 > media >= 0:
        print("Reprovado(a)!")
    
    else:
        print("Notas inválidas!")
```

### Função range()

**Formas de utilização:**
<ol>
    <li>range(stop);</li>
    <li>range(start, stop);</li>
    <li>range(start, stop, step);</li>
</ol>


```python
# 1ª Forma

intervalo = range(10)
print(*intervalo)

for i in range(10):
    print(i)

somador = 0
for i in range(100):
    somador = somador + i

print(somador)
```

```python
# 2ª Forma

for i in range(1, 11):
    print(i)
    
somador = 0
for i in range(50, 101):
    # somador = somador + i
    somador += i

print(somador)
```

```python
# 3ª Forma

for i in range(1, 20, 2):
    print(i)
    
somador = 0
for i in range(1, 101, 10):
    # somador = somador + i
    somador += i
    print(i)

print(somador)
```
