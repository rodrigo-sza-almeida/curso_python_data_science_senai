```python
# Exercício 1

numero = int(input('Digite o número: '))

# Iterações

for i in range(1, 11):
    print(f'{numero} X {i} = {i * numero}')
```

```python
# Exercício 2

quantidade_despesas = int(input("Quantidade de despesas"))
total_despesas = 0

for d in range(quantidade_despesas):
    valor_despesa = float(input(f'Valor da despesa {d+1}'))
    total_despesas += valor_despesa

print(f'Você gastou um total de: R$ {total_despesas:,.2f}')
```
