```python
palavras = [
    'projetor',
    'mouse',
    'garrafa',
    'caneta',
    'notebook',
    'teclado',
    'relógio',
    'controle',
    'boné'
]

palavras_maiusculas = list(map(lambda p: p.upper(), palavras))
print(palavras_maiusculas)

palavras_maiusculas_maiores = list(filter(lambda p: len(p) > 5, palavras_maiusculas))
print(palavras_maiusculas_maiores)
```

### Exercício 2

```python
from random import randint, seed

seed(1)
numeros = [randint(-20, 50) for _ in range(30)]
print(numeros)

numeros_positivos = list(filter(lambda x: x > 0, numeros))
print(numeros_positivos)

quadrados = list(map(lambda y: y ** 2, numeros_positivos))
print(quadrados)
```

### Exercício 3

```python
nomes = [
    'João Silva',
    'Ana Machado',
    'Daniel Sousa',
    'Maria Eduarda'
]

nomes_abreviados = list(
    map(
        lambda nome: f'{nome[0]}. {nome.split()[1][0]}.',
        nomes
    )
)

print(nomes_abreviados)
```

```python
for nome in nomes:
    print(nome.split()[0][0])
    print(nome.split()[1][0])
```
