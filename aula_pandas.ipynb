```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

d = {'a': 1, 'b': 2, 'c': 3}
ser = pd.Series(data=d, index=['a', 'b', 'c'])
```

```python
print(ser)
print(type(ser))
```

```python
ser.describe()
```

```python
# Acessar valores pelos 
print(ser['a'])
print(ser['b'])
print(ser['c'])
```
