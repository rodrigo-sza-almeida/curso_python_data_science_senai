### Objeto Date


```python
from datetime import date

hoje = date(2025, 12, 13)

print(hoje)
print(type(hoje))
```

```python
# Atributos

print(hoje.year)
print(hoje.month)
print(hoje.day)
```

```python
# Métodos

dia_atual = date.today()
print(dia_atual.year)
print(dia_atual.month)
print(dia_atual.day)

# Para mudar a representação
print(dia_atual.toordinal())  # Quantidade de dias desde 01/01/1970
print(dia_atual.isoweekday())

# Dia da semana
formatter = '%d/%m/%Y - %A'
print(dia_atual.strftime(formatter))

# Ou ainda

def date_formater(date: date) -> str:
    formatter = '%d/%m/%Y'
    return date.strftime(formatter)

data_nascimento = date(1998, 7, 11)
print(date_formater(data_nascimento))

print(date.max)
print(date.min)
```

### Objeto time

```python
from datetime import date, time

horario = time(10, 30, 10, 1234)
print(horario)
print(type(horario))

# Atributos
print(horario.hour)
print(horario.minute)
print(horario.second)
print(horario.microsecond)

# Métodos
print(time.max)
print(time.min)
print(horario.strftime('%H:%M:%S %p'))
```

```python
from datetime import datetime

agora = datetime.now()
print(agora)
print(type(agora))

# Atributos
print(agora.year)
print(agora.month)
print(agora.day)
print(agora.hour)
print(agora.minute)
print(agora.second)
print(agora.microsecond)

# Métodos
print(agora.strftime('%d/%m/%Y %H:%M:%S %p - %A'))
print(agora.astimezone())
print(datetime.utcnow())
```

### Objeto timedelta

```python
from datetime import timedelta, date, datetime

hoje = datetime.today()
print(hoje)
data_nascimento = datetime(1998, 7, 11, 9, 6)
dias_por_ai = hoje - data_nascimento
print(dias_por_ai)
print(type(dias_por_ai))

# Atributos e Métodos
print(dias_por_ai.days)
print(f'{dias_por_ai.total_seconds():,}')
print(timedelta.max)
print(timedelta.min)

dias_1000 = timedelta(1000)
print(hoje + dias_1000)
```

### Transformando uma data em texto para uma data objeto

```python
from datetime import date

dia_nascimento = '11/07/1998'

data = datetime.strptime(dia_nascimento, '%d/%m/%Y')
print(data)
print(type(data))
```
