### IF

Sintaxe:

<code>if condição:</code><br>
<code>    bloco do if</code>

```python
codigo_produto = "#10"
quantidade_produto = 1500

if quantidade_produto < 1000:
    print("Quantidade abaixo do mínimo.")
```

### IF - ELSE

Sintaxe:

<code>if condição:</code><br>
<code>    bloco do if</code><br>
<code>else:</code><br>
<code>    bloco do else</code>

```python
codigo_produto = "#10"
quantidade_produto = 500

if quantidade_produto < 1000:
    print("Quantidade abaixo do mínimo.")
    print(f"Comprar {1000 - quantidade_produto} unidades.")
else: 
    print(f"{quantidade_produto} está ok!")
```

### Usando Elif

Utilizaremos o Elif sempre que precisarmos de uma condição intermediário.

**Sintaxe:**

<code>if condição:</code><br>
<code><bloco do if</code><br>
<code>elif condição1:</code><br>
<code>    bloco do elif1</code><br>
...<br>
<code>elif condiçãoN</code><br>
<code>    bloco do elifN</code><br>
<code>else:</code><br>
<code>    bloco do else</code>

```python
codigo_produto = "#10"
quantidade_produto = 500

if quantidade_produto < 1000:
    print("Quantidade abaixo do mínimo.")
    print(f"Comprar {1000 - quantidade_produto} unidades.")

elif 1000 <= quantidade_produto <= 2500:
    print(f"{quantidade_produto} está ok!")

else:
    print("Quantidade acima do máximo!")
    print(f"Vender {quantidade_produto - 2500} uniddades.")
```

```python
# O que NÃO fazer!

codigo_produto = "#10"
quantidade_produto = 500

# Contexto 1
if quantidade_produto < 1000:
    print("Quantidade abaixo do mínimo.")
    print(f"Comprar {1000 - quantidade_produto} unidades.")

# Contexto 2
if 1000 <= quantidade_produto <= 2500:
    print(f"{quantidade_produto} está ok!")

else:
    print("Quantidade acima do máximo!")
    print(f"Vender {quantidade_produto - 2500} uniddades.")
```

```python
# Melhorando o nível de especificidade

codigo_produto = "#10"
quantidade_produto = 5500

# Contexto 1
if 1000 > quantidade_produto >= 0:
    print("Quantidade abaixo do mínimo.")
    print(f"Comprar {1000 - quantidade_produto} unidades.")

elif 1000 <= quantidade_produto <= 2500:
    print(f"{quantidade_produto} está ok!")

elif 2500 < quantidade_produto <= 5000:
    print("Quantidade acima do máximo!")
    print(f"Vender {quantidade_produto - 2500} uniddades.")

else:
    print('Quantidade inválida!')
```
