### Sets Python 

Utilização: Sempre que precisarmos de uma coleção ou série otimizada para estatística.

**Características:**
- Representação: {item1, item2, item3, ..., itemN}
- São iteráveis
- ❌ Não são indexadas em zero 
- Podem ser heterogêneas
- São editáveis ✏️
- Pertencem a classe set
- Não aceitam itens duplicados
- Não armazena informação de ordem

**Vantagens**
- Armazenamento de n informações
- Auxiliam no processamento de dados
- Sobretudo em análises estatísticas


```python
meu_set = {10, 3.14, 'oi', True}
print(meu_set)
print(type(meu_set))

for item in meu_set:
    print(item)
```

```python
numeros = set(range(51))

# Funções
print('Quanidade de itens:', len(numeros))
print('Soma dos itens:', sum(numeros))
print('Valor máximo:', max(numeros))
print('Valor mínimo:', min(numeros))
```

### Métodos de conjuntos

```python
s1 = {1, 2, 3, 4, 5}
s2 = {4, 5, 6, 7, 8}

# União
s_uniao = s1.union(s2)
print('Conjunto da união:', s_uniao)

# Interseção
s_interseccao = s1.intersection(s2)
print('Conjunto da Intersecção:', s_interseccao)

# Diferença
s1_diferenca = s1.difference(s2)
print('Conjunto da diferençe s1 em relação a s2:', s1_diferenca)
```
