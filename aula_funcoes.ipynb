### Funções ⛏️🔧🪚

Utilização: Sempre que precisarmos de um mini-software para resolver um problema dentro de nossa aplicação

**Uma função deve:**
- Ser específica para uma atividade, para ser excelente
- quanto mais simples melhor *
- bem documentada

**Vantagens 📈:**
- Reutilização de código
- Modularização de código
- Abstração
- Melhor organização do código

Sintaxe:

<code>def nome_funcao():</code><br>
<code>     o que a função faz</code>

```python
# Primeira função

# Declaração da função
def primeira_funcao():
    print('olá, função!')
```

```python
# Function call
primeira_funcao()
```

```python
def func():
    a = 10
    b = 10
    print(a + b)

func()
```

### Parâmetros

```python
def cumprimentar(nome):
    print(f'Oi, eu sou {nome}')

cumprimentar('João')
cumprimentar('Ana')
cumprimentar('Daniel')
cumprimentar('Maria')
```

```python
def soma(a, b):
    print(a + b)

soma(10, 10)
soma(2, 72)
print(soma(20, 15))
```

### Return x Tipo None

```python
def soma(a, b):
    return a + b
    print('oi')  # <- Essa linha é ignorada por vir depois de um return.

result = soma(10, 10)

print(result)
```

```python
def gera_email(nome, sobrenome):
    email = f'{nome}.{sobrenome}@senai.com'
    return email

print(gera_email('joao', 'silva'))
print(gera_email('ana', 'ferreira'))
print(gera_email('daniel', 'machado'))
```

### Escopo Global e Escopo Local

```python
a = 10

def f():
    global a
    a = 25

f()

print(a)
```

### Parâmetros de valor padrão

```python
def soma(a=0, b=0):
    print(f'a = {a} e b = {b}')
    soma = a + b
    return soma

result_1 = soma(10)
result_2 = soma(10, 10)
```

### Argumentos posicionais ou chave

```python
print(soma(1, 2))
print(soma(10, 50))
```

```python
print(soma(b=1, a=2))
print(soma(b=10, a=50))
```

### Typing Hint

```python
def soma(a: int, b: int) -> int:
    soma = a + b
    return soma
```

```python
result = soma(10, 10)
print(result)
```

```python
result = soma('10', '10')
```

```python
result
```

```python
def area_circunferencia(raio: int | float, precisao: int) -> int | float:
    """
    Função que recebe o raio e retorna o valor da área em m²
    :param raio: Raio com o qual desejamos calcular a área
    :return: Área em m²
    """
    from math import pi
    return round(pi * raio ** 2, precisao)
```

```python
print(area_circunferencia(1, 5))
print(area_circunferencia(2, 3))
```
