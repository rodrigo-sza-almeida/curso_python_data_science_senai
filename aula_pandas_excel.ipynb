### Análise das notas em excel

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Importanto a base de dados para o nosso df
df = pd.read_excel('notas.xlsx').copy()
```

```python
df.describe()
```

```python
# Calcular as médias para cada aluno
lista_medias = []

for i in range(len(df)):
    n1 = df.iloc[i]['N1']
    n2 = df.iloc[i]['N2']
    n3 = df.iloc[i]['N3']
    n4 = df.iloc[i]['N4']
    
    media = (n1 + n2 + n3 + n4) / 4
    print(df.iloc[i]['Nome do Aluno'], f'Média: {media}')
    
    lista_medias.append(round(media, 2))
    
print()
print(lista_medias)
```

```python
# Criando a coluna no DataFrame
df['Média'] = lista_medias
```

```python
df
```

```python
# Criar a coluna de Status no DataFrame

df['Status'] = [
    'Aprovado' if df['Média'][i] >= 5 else 'Reprovado' for i in range(len(df))
]
```

```python
df
```

```python
# Salvando so dados do DataFrame em Excel novamente
df.to_excel('notas_revisadas.xlsx')
```

```python
# Geração do gráfico
fig, ax = plt.subplots(figsize=(15, 5))

# Gerando uma lista com as cores para as barras do gráfico
cores = [
    'green' if df['Média'][i] >= 5 else 'red' for i in range(len(df))
]

g = ax.bar(df['Nome do Aluno'], df['Média'], color=cores, width=.5)

# Tornar o gráfico mais informativo.
ax.set_title('Médias Finais dos Alunos no ano de 2025', fontsize=15)  # Adicionando título
ax.set_xlabel('Nomes dos Alunos')  # Adicionando título ao eixo x
ax.set_ylabel('Méida dos Alunos')  # Adicionando título ao eixo y
ax.bar_label(g, padding=10)  # Adicionando os rótulos às barras
ax.set_ylim((0, 11))  # Ajustando a escala do eixo y
plt.show()
```
