### Operadores Matemáticos

**São eles:**
- + ➡️ Soma;
- - ➡️ Subtração;
- * ➡️ Multiplicação;
- / ➡️ Divisão;
- // ➡️ Divisão Inteira;
- % ➡️ Módulo (Resto da Divisão);
- ** ➡️ Potenciação;

Utilização: Sempre que precisarmos realizar um cálculo específico ou criar uma<br> 
fórmula customizada, por exemplo para cálculo de juros compostos.

🤔 Regra de Sinais:

\+ com + ➡️ + <br>
\+ com - ➡️ - <br>
\- com + ➡️ - <br>
\- com - ➡️ + <br>

🚩 Prioridades:
1 - (  )
2 - Potenciação
3 - Divisões e Multiplicações
4 - Somas e Subtrações


```python
# Soma

print(10 + 10)
print(10 + -10)
print(-10 + 10)
print(-10 + -10)
print()

# Subtração

print(10 - 10)
print(10 - -10)
print(-10 - 10)
print(-10 - -10)
print()

# Multiplicação

print(10 * 10)
print(10 * -10)
print(-10 * 10)
print(-10 * -10)
print(10 + 10 * 3)
print((10 + 10) * 3)
print()

# Divisão

print(10 / 10)
print(10 / -10)
print(-10 / 10)
print(-10 / -10)
print(10 + 10 / 3)
print((10 + 10) / 3)
print()

# Potenciação

print(10 ** 2)
print(10 ** -2)
print(-10 ** 2)
```

### Divisão Inteira (//) e Módulo - Resto da Divisão (%)

```python
print(10 / 2)
print(10 // 2)
print(10 % 2)
print()
print(10 / 3)
print(10 // 3)
print(10 % 3)
print(20 // 3)
print(20 / 3)
```

```python
# Exemplo Prático 

area = 1000
litros = area / 15

n_latas_18 = litros // 18
n_latas_3 = (litros % 18) // 3.6

print(f"Área: {area:,} m²")
print(f"Quantidade de latas de 18L: {n_latas_18}")
print(f"Quantidade de latas de 3,6L: {n_latas_3}")
```
