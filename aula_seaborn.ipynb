```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()

# Geração do valores
np.random.seed(42)
x = np.arange(1, 11)
a = np.random.randint(1, 100, 10)
b = np.random.randint(1, 100, 10)

# Separando imagem e eixos
fig, ax = plt.subplots(figsize=(15, 5))

# Curva A
ax.plot(x, a, color='#582770', linewidth=3, marker='d', markersize=7, alpha=.3,
       label='Empresa A')

# Curva B
ax.plot(x, b, color='#2eb8ac', linewidth=3, marker='o', markersize=7,
       label='Empresa B')

# ax.set_title('Ações da empresa x', fontsize=15) --> Para alterar o tamanho da fonte
ax.set(
    title='Ações empresa A',
    xlabel='Meses do Ano',
    ylabel='Valor da ação em R$',
    ylim=(0, 100),
    xlim=(0, 11)
)
ax.legend()
ax.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

vendas_produtos = {
    'A': 100,
    'B': 250,
    'C': 175,
    'D': 230,
    'E': 270
}

# Lista de cores
# cores = ['#484848', '#006465', '#0f928c', '#00c9d2', '#beee3b']

# Gerando imagem e eixos(axes)
fig, ax = plt.subplots(figsize=(15, 5))

# Criado e armazenando informações do gráfico de barra para os rótulos
g = ax.bar(vendas_produtos.keys(),
           vendas_produtos.values(), width=.5, color=sns.mpl_palette('Set2'))
ax.set(
    title='Vendas por Produto no ano de 2025',
    xlabel='Nome do Produto',
    ylabel='Valor das vendas em Milhões de R$',
    ylim=(0, 320)
)
ax.bar_label(g, padding=10)
plt.show()
```
