### Expressões Lambda: As funções anônimas do Python

Utilização: Sempre que precisarmos de uma função leve e otimizada, que possa ser utilizada como função de callback, por exemplo:

- Precisamos aplicar um filtro, a determinação desse filtro seria um callback.

**Vantagens:**
- Função otimizada
- Não será armazenada em memória, o que pode contribuir para o desempenho
- Podem ser usadas como 'callback'

Sintaxe:
lambda x: f(x)

```python
# Exemplo 1

quadrado = lambda x: x ** 2
```

```python
quadrado
```

```python
print(quadrado(10))
print(quadrado(5))
print(quadrado(7))
```

```python
abreviar_nome = lambda nome, sobrenome: nome.upper()[0] + '. ' + sobrenome.upper()[0] + '.'
```

```python
print(abreviar_nome('joão', 'silva'))
print(abreviar_nome('ana', 'ferreira'))
print(abreviar_nome('daniel', 'machado'))
```

```python
par = lambda x: x % 2 == 0
```

```python
print(par(1))
print(par(2))
print(par(3))
print(par(4))
```

### Função map()

```python
numeros = [1, 2, 3, 4, 5]

digitos = list(map(str, numeros))
print(digitos)

def quadrado(n):
    return n ** 2

numeros_quadrados = list(map(quadrado, numeros))
print(numeros_quadrados)
```

```python
numeros = [1, 2, 3, 4, 5]

numeros_quadrados = list(map(lambda n: n ** 2, numeros))
```

```python
numeros_quadrados
```

### Função filter()

```python
numeros = [1, 2, 3, 4, 5]

def par(n):
    return n % 2 == 0
    
pares = list(filter(par, numeros))
print(pares)

maiores_que_3 = list(filter(lambda x: x > 3, numeros))
print(maiores_que_3)
```

```python
print(3 & 5, bin(3), bin(5))
```

```python
print(3 | 5, bin(3), bin(5))
```
