### Tuplas Python 🗄️

Utilização: Sempre que precisarmos de uma coleção ou série para armazenar mais de um dado ou informação, que seja constante.

**Características:**
- Representação: (item1, item2, item3, ..., itemN)
- São iteráveis
- São indexadas em zero 
- Podem ser heterogêneas
- ❌ Não são editáveis
- Pertencem a classe tuple

**Vantagens**
- Armazenamento de n informações
- Tuple unpacking

```python
minha_tupla = (10, 3.14, 'oi', True)
print(minha_tupla)
print(type(minha_tupla))

for item in minha_tupla:
    print(item)
```

### Indexação em zero

```python
numeros = tuple(range(51))

# Index
print(numeros[0])
print(numeros[5])
print(numeros[30])
print(numeros[-1])

# SLicing
print(numeros[0:10])
print(numeros[10:20])
print(numeros[:20])
print(numeros[30:])
print(numeros[::5])
print(numeros[::-1])

# Funções
print('Quanidade de itens:', len(numeros))
print('Soma dos itens:', sum(numeros))
print('Valor máximo:', max(numeros))
print('Valor mínimo:', min(numeros))
```

```python
# Tuplas não podem ser alteradas em tempo de execução

import datetime as dt

minha_tupla = ('Thiago', 'thiago@gmail.com', 27, dt.date(1998, 7, 11))

# minha_tupla[0] = 'Pedro'
```

### Tupla Unpacking

```python
nome, email, idade, data_nascimento = minha_tupla
```

```python
print(nome)
print(email)
print(idade)
print(data_nascimento)
```
