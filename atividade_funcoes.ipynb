### Exercício 1:
Crie uma função que receba uma lista com números como argumento e retorne uma nova lista apenas com os valores pares da original.

```python
from typing import List
import random

random.seed(42)
numeros = [random.randint(0, 1000) for _ in range(100)]  # Criando uma lista de aleatórios

def filtra_pares(numeros: List) -> List:
    """
    Função que filtra os valores pares de uma liats númerica e retorna
    """
    pares = []
    for n in numeros:
        if n % 2 == 0:
            pares.append(n)
        else:
            pass
    return pares

pares = filtra_pares(numeros)
print(pares)
```

```python
from typing import List
import random

random.seed(42)
numeros = [random.randint(0, 1000) for _ in range(20)]

def filtra_pares(numeros: List = []) -> List:
    """
    Função que filtra os valores pares de uma liats númerica e retorna
    """
    return [n for n in numeros if n % 2 == 0]

print(filtra_pares(numeros))
```

### Exercício 2
Crie uma função que receba um número inteiro (até 999) e diga a composição do número em centenas, dezenas e unidades, por exemplo:
Entrada: 326
Saída: 3 centenas, 2 dezenas e 6 unidades.

```python
def decompor_numero(numero: int) -> None:
    """
    Função que decompões um número (até 999) e mostra.
    """
    centenas = numero // 100
    dezenas = numero % 100 // 10
    unidades = numero % 10
    print(f'{centenas} centenas - {dezenas} dezenas - {unidades} unidades')
```

```python
decompor_numero(326)
```

```python
decompor_numero(751)
```

### Exercício 3
Crie uma função que receba um número inteiro e informe se ele é primo ou não.

```python
# Um número é primo se ele é divisível por 1 e por ele mesmo
# Se um número não é divisível pelos números até a metade de seu valor
# ele é primo

def primo(numero: int) -> None:
    """
    Função que recebe um valor e diz se ele e primo ou não
    """
    if numero == 2 or numero == 3:
        print(f'{numero} é primo')
        return
    for n in range(2, numero // 2 + 1):
        if numero % n == 0:
            print(f'{numero} não é primo')
            return 
    print(f'{numero} é primo')
```

```python
for i in range(2, 100):
    primo(i)
```
