### Exercício 1

```python
# Declarar as variáveis

n1 = float(input("N1: "))
n2 = float(input("N2: "))

# Cálculo da média

media = (n1 + n2) / 2

print()
print(f"Sua média é de {media:.2f}")

# Condicional

if media == 10:
    print("Aprovado(a) com distinção!")

elif 10 > media >= 6:
    print("Aprovado(a)!")

elif 6 > media >= 0:
    print("Reprovado(a)!")

else:
    print("Notas inválidas!")

```

### Exercício 2

```python
# Declarar as variáveis

salario_bruto = float(input("Salário: R$"))
desconto = 0

# Condicionais
if salario_bruto >= 0:
    if salario_bruto <= 2000:
        print("Isento.")

    elif 2000 < salario_bruto <= 3500:
        desconto = .1

    elif 3500 < salario_bruto <= 5000:
        desconto = .15

    elif salario_bruto > 5000:
        desconto = .2

    salario_liquido = salario_bruto * (1 - desconto)

    print()
    print(f'Salário bruto: R$ {salario_bruto:,.2f}')
    print(f'Desconto: {desconto:.2%}')
    print(f'Valor descontado: R$ {desconto * salario_bruto:,.2f}')
    print(f'Salário líquido: R$ {salario_liquido:,.2f}')
else:
    print(f'O valor R$ {salario_bruto:,.2f} é inválido!')
```

### Operações ternárias

```python
# Condicionais me uma linha apenas

media = 4

status_aluno = 'Aprovado' if media >= 5 else 'Reprovado'
print(f'Situação do aluno: {status_aluno}')

# Exemplo 2

numero = 11

funcao = numero * 10 if numero % 2 == 0 else numero / 10
print(f'Resultado: {funcao}')
```

### Match / Case

```python
# Exmemplo 1

dia = int(input("Digite o dia da semana (1 - domingo): "))

match dia:
    case 1:
        print('Domingo')
    case 2:
        print('Segunda-feira')
    case 3:
        print('Terça-feira')
    case 4:
        print('Quarta-feira')
    case 5:
        print('Quinta-feira')
    case 6:
        print('Sexta-feira')
    case 7:
        print('Sábado')

    case _:
        print('Dia inválido!')
```
