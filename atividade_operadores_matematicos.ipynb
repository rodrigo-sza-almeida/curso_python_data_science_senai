### Exercício 1

```python
import math

# Declaração das variáveis

pi = math.pi
raio = int(input("Raio em m:"))

# Calcular as grandezas

diametro = 2 * raio
perimetro = 2 * pi * raio
area = pi * raio ** 2

# Exibição dos valores
print()
print(f"Diâmetro: {diametro:.4f} m")
print(f"Perímetro: {perimetro:.4f} m")
print(f"Área: {area:.4f} m²")
```

### Exercício 2

```python
# Declaração de variáveis

temp_celsius = 20

# Converter para °F
temp_fahrenheit = 1.8 * temp_celsius + 32
print(f"{temp_celsius}°C equivalem a {temp_fahrenheit}°F")


# Converter para K
temp_fahrenheit = temp_celsius + 273.15
print(f"{temp_celsius}°C equivalem a {temp_fahrenheit}K")
```

### Exercício 3


```python
from statistics import mean

n1 = float(input("N1: "))
n2 = float(input("N2: "))
n3 = float(input("N3: "))
n4 = float(input("N4: "))

# Calcular a média
# soma = n1 + n2 + n3 + n4
# media = soma / 4
media = mean([n1, n2, n3, n4])

print(f"Sua média final é de: {media:.2f}")
```
