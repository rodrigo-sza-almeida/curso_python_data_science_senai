### Strings

Utilização: Sempre que precisarmos de um dado que se comporte como texto.

**Características / Vantagens:**
- Simples porém possuem muitos métodos para nos auxiliar;
- Compatível com inúmeras tecnologias (liguagens, Bancos de Dados);
- São iteráveis;
- São indexadas em zero;

### Métodos de String

```python
texto = "Python para data science"

# Métodos de formatação
print(texto.upper())
print(texto.lower())
print(texto.title())
print(texto.capitalize())
print(texto.replace("a", "A"))
print(texto.replace("a", "A", 1))  # Quando você quer mudar apenas uma quantidade de ocorrências
print(texto.split())
print()
# Métodos de validação (retornam True ou False)

codigo_produto = "ELT10"
print(f"O produto {codigo_produto} é da área de elétrica: {codigo_produto.startswith("ELT")}")

email = 'joao.silva@senai.com'
print(f"{email} é um email válido: {email.endswith("@senai.com")}")
nome = 'joao'
print(nome.isalpha())

telefone = '11999999999'
print(telefone.isdigit())

codigo_produto = "ELT10 "
print(codigo_produto.isalnum())

print(texto.__contains__('a'))

# Métodos úteis

# Contar quantas vezes dada string aparece em uma outra
print(texto.count('a'))

# Encontrar a posição de um strings dentro de uma outra
print(texto.index('y'))

# Para descobrir a quantidade de caracteres de uma string
print(len(texto))
```

```python
from collections import Counter
```

```python
print(Counter(texto))
```

### Indexação de Strings

```python
# Índice
texto = "Python para data science"

print(texto[0])
print(texto[1])
print(texto[23])
print(texto[-1])
# print(texto[50]) --> IndexError
```

```python
# Slicing

texto = "Python para data science"

print(texto[0:5])
print(texto[5:10])
print(texto[15:20])
print(texto[20:30])
print(texto[:10])
print(texto[10:])
print(texto[::5])
print(texto[::-1])

telefone = "11999999999"

estado = "SP" if telefone[0:2] == "11" else "Outro"
print(estado)
```
