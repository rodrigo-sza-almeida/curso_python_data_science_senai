### Manipulando JSON { }

JSON significa JavaScript Object Notation e se tornou praticamente um padrão para as APIs (Application Programming Interface).

Arquivos .JSON são arquivos de texto, que carregam dados no formato semi-estruturado.

Utilizando as APIs, podemos obter JSON como output, com as informações pertinentes a nossas requests (requisições)

Para praticarmos, utilizaremos a API de CEP do AwesomeAPI:

<a href='https://docs.awesomeapi.com.br/api-cep'>AwesomeAPI CEP</a>

```python
import requests

cep = '09751-000'

req = requests.get(f'https://cep.awesomeapi.com.br/json/{cep}')
```

```python
informacoes_cep = req.json()
```

```python
print(f'Logradouro: {informacoes_cep['address']}')
print(f'Bairro: {informacoes_cep['district']}')
print(f'Cidade: {informacoes_cep['city']}')
print(f'Estado: {informacoes_cep['state']}')
```
