### Gráficos de Dispersão

#### Ideais para representar o comportamento de 2 variáveis numéricas

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)

x = np.random.randint(1, 1000, 100)
y = np.random.randint(1, 1000, 100)

plt.scatter(x, y, c='#FFAACC', marker='h', alpha=.3, s=100)
plt.title('Para 2 grandezas numéricas')
plt.show()
```

### Gráfico de Linha

#### Ideal para analisar uma grandeza ao longo do tempo

```python
import matplotlib.pyplot as plt
import numpy as np

# Geração do valores
np.random.seed(42)
x = np.arange(1, 11)
y = np.random.randint(1, 100, 10)

# Separando imagem e eixos
fig, ax = plt.subplots(figsize=(15, 5))
ax.plot(x, y, color='#582770', linewidth=3, marker='d', markersize=7, alpha=.3)
# ax.set_title('Ações da empresa x', fontsize=15)
ax.set(
    title='Ações empresa A',
    xlabel='Meses do Ano',
    ylabel='Valor da ação em R$',
    ylim=(0, 100),
    xlim=(0, 11)
)
ax.grid()
plt.show()
```

### Gráfico de linhas sobrepostas

```python
import matplotlib.pyplot as plt
import numpy as np

# Geração do valores
np.random.seed(42)
x = np.arange(1, 11)
a = np.random.randint(1, 100, 10)
b = np.random.randint(1, 100, 10)

# Separando imagem e eixos
fig, ax = plt.subplots(figsize=(15, 5))

# Curva A
ax.plot(x, a, color='#582770', linewidth=3, marker='d', markersize=7, alpha=.3,
       label='Empresa A')

# Curva B
ax.plot(x, b, color='#2eb8ac', linewidth=3, marker='o', markersize=7,
       label='Empresa B')

# ax.set_title('Ações da empresa x', fontsize=15) --> Para alterar o tamanho da fonte
ax.set(
    title='Ações empresa A',
    xlabel='Meses do Ano',
    ylabel='Valor da ação em R$',
    ylim=(0, 100),
    xlim=(0, 11)
)
ax.legend()
ax.grid()
plt.show()
```

### Gráficos de Barra 

#### Ideal para mostrar valores por categorias ou por rótulos

```python
import matplotlib.pyplot as plt

vendas_produtos = {
    'A': 100,
    'B': 250,
    'C': 175,
    'D': 230,
    'E': 270
}

# Lista de cores
cores = ['#484848', '#006465', '#0f928c', '#00c9d2', '#beee3b']

# Gerando imagem e eixos(axes)
fig, ax = plt.subplots(figsize=(15, 5))

# Criado e armazenando informações do gráfico de barra para os rótulos
g = ax.bar(vendas_produtos.keys(),
           vendas_produtos.values(), width=.5, color=cores)
ax.set(
    title='Vendas por Produto no ano de 2025',
    xlabel='Nome do Produto',
    ylabel='Valor das vendas em Milhões de R$',
    ylim=(0, 320)
)
ax.bar_label(g, padding=10)
plt.show()
```

```python
import matplotlib.pyplot as plt

vendas_produtos = {
    'A': 100,
    'B': 250,
    'C': 175,
    'D': 230,
    'E': 270
}

# Lista de cores
cores = ['#484848', '#006465', '#0f928c', '#00c9d2', '#beee3b']

# Gerando imagem e eixos(axes)
fig, ax = plt.subplots(figsize=(15, 5))

# Criado e armazenando informações do gráfico de barra para os rótulos
g = ax.barh(vendas_produtos.keys(),
           vendas_produtos.values(), color=cores)
ax.invert_yaxis()
ax.set(
    title='Vendas por Produto no ano de 2025',
    xlabel='Nome do Produto',
    ylabel='Valor das vendas em Milhões de R$'
)
ax.bar_label(g, padding=10)
plt.show()
```
