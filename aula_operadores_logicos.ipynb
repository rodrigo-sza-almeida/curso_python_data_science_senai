### Operadores Lógicos

Esses operadores tem origem na Matemática / Lógica Booleana.<br>

**São eles:**
- and (E);
- or (OU);
- not (Negação);

Utilização: Sempre que precisarmos estruturar condicionais ou expressões lógicas mais complexas.<br>

*Exemplo*: Quando precisarmos validar o usuário E a senha para autenticar o usuário.

**Curiosidade**: Esses operadores também são a base dos sistemas digitais, sendo fisicamente implementados como circuitos eletrônicos integrados aos dispositivos que utilizamos como celular, computador entre outros.



```python
# AND (E)

print(False and False)
print(False and True)
print(True and False)
print(True and True)
print()

print(10 == 10 and 100 % 2 == 0)
print()

# Exemplo prático

user = 'admin'
password = 'adm'
auth = user == 'admin' and password == 'admin'

print(auth)
```

```python
# OR (OU)

print(False or False)
print(False or True)
print(True or False)
print(True or True)
print()

print(10 == '10' or 10 > 1)
print()

# Exemplo de aplicação

media = 10
freq = .85

reprova_aluno = media < 5 or freq < .75

print(f"O aluno vai reprovar: {reprova_aluno}")
```

```python
# not

print(not False)
print(not True)
print()

print(not(10 == 11 and 10 > 2 or 4 > 2))

# Exemplo prático

C = False
B = True
A = True

bomba = \
(not A and not B and not C) or (A and not B and not C) or (A and B and not C)

print(f"Liga a a bomba: {bomba}")
```
