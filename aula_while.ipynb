### Loops While 🔁

Utilização: Sempre que precisarmos de uma estrutura de repetição que esteja codicionado a um estado da aplicação. É correto afirmar que Loops while são para repetições condicionais.

**Vantagens:**
- Otimização de código;
- Maior controle de repetições do código;

**Sintaxe**

<code>while condição:</code><br>
<code>    bloco while</code><br>

⚠️ Aviso: Cuidado com loops infinitos

```python
i = 0

while i < 10:
    print(i)
    i += 1 
```

```python
salario_bruto = float(input("Salário"))
desconto = 0

while salario_bruto < 0:
    print('Salário inválido!')
    salario_bruto = float(input("Salário"))
else:
    if salario_bruto <= 2000:
        print("Isento.")
    
    elif 2000 < salario_bruto <= 3500:
        desconto = .1
    
    elif 3500 < salario_bruto <= 5000:
        desconto = .15
    
    elif salario_bruto > 5000:
        desconto = .2
    
    salario_liquido = salario_bruto * (1 - desconto)
    
    print()
    print(f'Salário bruto: R$ {salario_bruto:,.2f}')
    print(f'Desconto: {desconto:.2%}')
    print(f'Valor descontado: R$ {desconto * salario_bruto:,.2f}')
    print(f'Salário líquido: R$ {salario_liquido:,.2f}')
```

### continue e break

```python
# Exemplo de Login

while True:

    user = input('user')

    if user == 'admin':
        password = input('password')

        if password == 'admin':
            print('auth')
            break

        else:
            print('senha inválida')
            continue

        
```

```python
def banco():
    # Sistema de Banco

    saldo = 0
    
    print('--- Sistema Banco Senai 🏦 ---')
    
    while True:
    
        print(f'Seu saldo final é de R$ {saldo:,.2f}')
        
        print('''
        Escolha a operação que deseja:
        [ 1 ] -> Depósito
        [ 2 ] -> Saque
        [ 3 ] -> Sair
        ''')
        operacao = input(':')
    
        match operacao:
        
            case '3':
                print('Volte sempre ...')
                break # para sair do loop ❌
        
            case '1':
                print('--- Depósito ---')
                valor = float(input('Valor a ser depositado: R$ '))
                saldo += valor  # incrementando de forma elegante 👌
        
            case'2':
                print('--- Saque ---')
                valor = float(input('Valor a ser sacado: R$ '))
                if valor <= saldo:
                    saldo -= valor # decrementando de forma elegante 👌
                else:
                    print('Saldo insuficiente!')
                    continue
            
            case _:
                print('Operação inválida!')
                continue
    print(f'Seu saldo final é de R$ {saldo:,.2f}')
        
tentativas = 0

while True:

    if tentativas < 3:

        user = input('user')
    
        if user == 'admin':
            password = input('password')
    
            if password == 'admin':
                print('auth')
                banco()
                break
    
            else:
                print('senha inválida')
                continue
    else:
        print('Número de tentativas excedido!')

!pip install jupyterlab-indent-guides   
```
