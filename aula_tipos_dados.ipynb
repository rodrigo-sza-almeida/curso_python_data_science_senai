### Tipos de dados

**Os primeiros tipos de dados:**
- int: Para números inteiros;
- float (ponto flutuante): Para números com casas decimais;
- str (string): Para qualquer texto;
- bool (boolean): Valores lógicos, apenas True ou False;

### int

```python
print(10, type(10))
print(-10, type(-10))
print(100000, type(100000))
print(-100000, type(-100000))
```

### float

```python
print(10.0, type(10.0))
print(3.14, type(3.14))
print(100000.123, type(100000.123))
print(-100000.123, type(-100000.123))

b = 3.14.
```

### Strings

```python
print("olá mundo!", type("olá mundo!"))
print('olá mundo!', type('olá mundo!'))
print("senha123", type("senha123"))
print("12345", type("12345"))
print("!@#$#$¨&", type("!@#$#$¨&"))
print("print('oi')", type("print('oi')"))

print("olá 'mundo'")
print('olá "mundo"')

print("olá \"mundo\"")
```

### bool (Boolean)

```python
print(True, type(True))
print(False, type(False))
```
