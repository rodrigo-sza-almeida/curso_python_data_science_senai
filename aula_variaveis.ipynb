### Variáveis

Utilização: Em praticamente todo código precisamos de variáveis, para armazenar informações.<br>

**Vantagens:**
- Armazenam dados;
- Tornam o código mais dinâmico;
- Auxiliam no processamento de dados;
- Organizam as informações;

**Sintaxe:**
<code>nome_variavel = valor</code>

```python
a = 10
```

```python
a + 10 * 10
```

```python
id(a)
```

❌ Regras ❌

- Nomes de variáveis não podem iniciar com números: 1berto;
- Nomes de variáveis não podem ter char especiais: !@#$%¨&*()-, Exemplo: *senha;
- Não pode receber o nome de uma palavra reservada python: Exemplo, if, else, in, as, import, etc
- Não pode ter espaço no nome da variável;


```python
1berto = True
```

```python
quantidade* = 100
```

```python
if = "Oi"
```

```python
from keyword import kwlist
print(kwlist)
```

🌏 Boas práticas 🌏

- O padrão para os nomes de variáveis em Python é o snake_case: nome_aluno;
- Sempre com letras minúsculas: nome_aluno;
- Nomear a variável de forma a indicar muito claramente a que ela se refere: nome_aluno;

⚠️ OBS: Python é uma linguagem *Case Sensitive*, que quer dizer, diferencia letras maiúsculas de minúsculas.<br>
Exemplo: quantidade é diferente de Quantidade

```python
quantidade_produto = 1000
Quantidade_produto = 2000
```

```python
print(quantidade_produto, id(quantidade_produto))
print(Quantidade_produto, id(Quantidade_produto))
```
